```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df = pd.read_csv("C:/Users/kanha/archive (1)/netflix_titles.csv")
```

```python
df.head()
```

```python
df["description"] =df["description"].fillna("")
```

```python
df.head()
```

```python
df.description
```

```python
tf_idf = TfidfVectorizer(stop_words="english")
```

```python
print(tf_idf)
```

```python
tf_idf_metrix = tf_idf.fit_transform(df["description"])
```

```python
print(tf_idf_metrix)
```

```python
tf_idf_metrix.shape
```

```python
cosine_sim = linear_kernel(tf_idf_metrix, tf_idf_metrix)
```

```python
indices = pd.Series(df.index, index= df["title"]).drop_duplicates()
```

```python
print(indices)
```

```python
#now lets creat recomendtion function
```

```python
def get_recoommendations(title, cosine_sim=cosine_sim, num_recommend=10):
    idx =indices[title]
    sim_scores = list (enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar = sim_scores[1:num_recommend+1]
    movie_incides = [i[0] for i in top_similar]
    return df["title"].iloc[movie_incides]
```

```python
get_recoommendations("Blood & Water", num_recommend=10)
```
